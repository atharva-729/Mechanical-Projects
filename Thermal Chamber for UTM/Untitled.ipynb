```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tomake = ["LSR 50%.csv", "LSR Room temp..csv", "LSR70%.csv", "pla 50%.csv", "pla room temp..csv", 
         "room temp pla 2.csv", "rubber 50%.csv", "rubber70%.csv"]

df_dict = {}
for x in tomake:
    df_dict[x] = pd.read_csv(x)
```

```python
df_dict
```

```python
for key,value in df_dict.items():
    df = pd.DataFrame(columns= value.columns)
    for i in np.arange(0 ,len(value) , len(value)/30 ) :
        
        df.loc[int(i)] = value.loc[int(i)]
        df.to_csv(f"modified_{key}")
```
